```python
from sklearn.linear_model import LinearRegression 
x = data[['walkDistance']] 
y = data['winPlacePerc']
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)
from sklearn.linear_model import LinearRegression 
mlr= LinearRegression() 
mlr.fit(x_train, y_train)
print('winPlacePerc=%2f+walkDistance*%2f' %(mlr.intercept_,mlr.coef_))
```

```python
y_pred = mlr.predict(x_test)
plt.scatter(x_train, y_train,color='g') 
plt.xlabel('walkDisatnce')
plt.title('Linear regression of walkdistance and winPlaceperc')
plt.ylabel('winPlacePerc')

plt.plot(x_test, y_pred,color='k')
mlr.score(x,y)
```

```python
import matplotlib.pyplot as plt

plt.style.use('default')
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(7, 3.5))

ax.plot(x_test, y_pred, color='k', label='Regression model')
#ax.scatter(x, y, edgecolor='k', facecolor='grey', alpha=0.7, label='Sample data')
ax.set_ylabel('winplacePer', fontsize=14)

ax.legend(facecolor='white', fontsize=11)
ax.text(0.55, 0.15, '$y = %.2f  + %.2fx_1+%2fx_2+%2fx_3 $' % ( abs(mlr.intercept_),mlr.coef_[0],mlr.coef_[1],mlr.coef_[2]), fontsize=17, transform=ax.transAxes)

fig.tight_layout()
```
