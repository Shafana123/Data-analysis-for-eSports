```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
data=pd.read_excel('PUBG - Squad (352k rows).xlsx')
data
```

```python
data.info()
```

```python
f=data['weaponsAcquired']
g=data['matchDuration']
h=data['kills']

fig, ax = plt.subplots()

scatter = ax.scatter(f, g, c=h)

# produce a legend with the unique colors from the scatter
legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper right", title="kills")
ax.add_artist(legend1)

# produce a legend with a cross-section of sizes from the scatter
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)

plt.xlabel('No of weapons')
plt.ylabel(' Match Duration(s)')
plt.title('Weapons v/s match duration')
plt.show()
```

```python
columns1=['weaponsAcquired','matchDuration','kills']
selected_data=data[columns1]
corr_matrix = selected_data.corr()
sn.heatmap(corr_matrix, annot=True)
plt.title('correlation heatmap')
plt.show()
```

```python
columns1=['weaponsAcquired','matchDuration','kills','winPlacePerc']
selected_data=data[columns1]
corr_matrix = selected_data.corr()
sn.heatmap(corr_matrix, annot=True)
plt.title('correlation heatmap')
plt.show()
```

```python
g = sn.pairplot(data, vars=["weaponsAcquired", "kills","winPlacePerc"])
plt.show()
```
